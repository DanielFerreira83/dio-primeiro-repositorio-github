```python
# Criando nosso Dataframe
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
# Visualizando 5 primeiras linhas
df.head()
```

```python
# Número de linha e colunas
df.shape
```

```python
# VErificando os tipos de dados
df.dtypes
```

```python
# Qual receita total
df['Valor Venda'].sum()
```

```python
# Qual o custo total
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade']) # cria coluna de custo
```

```python
df.head()
```

```python
# Custo total
round(df['Custo'].sum(),2)
```

```python
# Agora que temos a receita o custo e o total, podemos achar o lucro total
# Vamos criar coluna de lucro que será Receita - Custo
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head(1)
```

```python
# Total lucro
round(df['Lucro'].sum(),2)
```

```python
# Criando uma coluna com Total de dias para enviar o produto
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

#### Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo Envio em numérico

```python
# Encontrando apenas os dias
df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
# VErificando tipo da coluna Tempo Envio
df['Tempo Envio'].dtypes
```

```python
# Média do tempo de envio por Marcas (em dias)
round(df.groupby('Marca')['Tempo Envio'].mean(),2)
```

```python
# Qual total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Gráfico total produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total Produtos Vendidos')
plt.xlabel("Total")
plt.ylabel('Produto')
```

```python
# Lucro por ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro por ano')
plt.xlabel('Ano')
plt.ylabel('Receita')
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
# Selecioando vendas do ano de 2019
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
# gráfico de linhas lucro por mes
df_2009.groupby(df['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucros por mês')
plt.xlabel('Mês')
plt.ylabel('Lucro')
```

```python
# gráfico de barras lucro por marca
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucros por Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal')
```

```python
# gráfico de barras lucro por classe
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro por Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal')
```
